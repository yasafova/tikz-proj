```python
import os
import shutil

# Define source directories
dir1 = 'arvix/content/codes/pngs'  # Replace with your actual path
dir2 = 'final/pngs'  # Replace with your actual path

# Define destination directory
combined_dir = 'dataset/pngs'
os.makedirs(combined_dir, exist_ok=True)

def copy_all(src, dst):
    for item in os.listdir(src):
        src_path = os.path.join(src, item)
        dst_path = os.path.join(dst, item)

        if os.path.isdir(src_path):
            if not os.path.exists(dst_path):
                shutil.copytree(src_path, dst_path)
            else:
                # Recursively merge subdirectories
                copy_all(src_path, dst_path)
        else:
            if not os.path.exists(dst_path):
                shutil.copy2(src_path, dst_path)

# Copy from both directories
copy_all(dir1, combined_dir)
copy_all(dir2, combined_dir)

print(f"Contents of {dir1} and {dir2} have been merged into {combined_dir}")
```

```python
import pandas as pd

# Load your CSV file
df = pd.read_csv('stackfinal/tikz_dataset.csv')  # Replace with your actual file path

# Function to generate the new 'source' value
def generate_source_value(row):
    question_value = row['question_id']
    answer_value = row['answer_id'] if pd.notnull(row['answer_id']) else "None"
    return f"question_id: {question_value}, answer_id: {answer_value}"

# Apply the function to generate the 'source' column
df['source'] = df.apply(generate_source_value, axis=1)

# Save the updated DataFrame
df.to_csv('stackfinal/tikz_dataset2.csv', index=False)

print("The 'source' column has been updated.")
```
